```python
#4
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt  


data = iio.imread('fruit.jpg', mode='F') 
print(data.shape)


d1 = nd.rotate(data, 20)
plt.imshow(d1, cmap='gray')
plt.title('Xoay 20 độ với reshape=True')
plt.show()


d2 = nd.rotate(data, 20, reshape=False)
plt.imshow(d2, cmap='gray')
plt.title('Xoay 20 độ với reshape=False')
plt.show()
```

```python
#5
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
from skimage.color import rgb2gray
from skimage.util import img_as_ubyte

data = iio.imread('world_cup.jpg')
print(data.shape)  
gray = rgb2gray(data)
binary = gray > 0.5  
d1 = nd.binary_dilation(binary)
plt.imshow(d1, cmap='gray')
plt.title('Dilation 1 iteration')
plt.show()

d2 = nd.binary_dilation(binary, iterations=3)
plt.imshow(d2, cmap='gray')
plt.title('Dilation 3 iterations')
plt.show()
```

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('world_cup.jpg')

from skimage.color import rgb2gray
gray = rgb2gray(data)

structure = np.ones((5, 5), dtype=bool)

d1 = nd.binary_dilation(gray > 0.5, structure=structure)
plt.imshow(d1, cmap='gray')
plt.title('Dilation với size=(5, 5)')
plt.show()
```

```python
#6
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
from skimage.color import rgb2gray
img_color = iio.imread('world_cup.jpg')
data = rgb2gray(img_color)

print(data.shape)
V, H = data.shape
M = np.indices((V, H))
d = 5
q = 2 * d * np.random.ranf(M.shape) - d
mp = (M + q).astype(int)
dl = nd.map_coordinates(data, mp)
plt.imshow(dl, cmap='gray')
plt.title('Hiệu ứng nhiễu tọa độ')
plt.show()
```

```python
#7
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt
from skimage.color import rgb2gray

def GeoFun(outcoord):
    a = 10 * np.cos(outcoord[0] / 10.0) + outcoord[0]
    b = 10 * np.cos(outcoord[1] / 10.0) + outcoord[1]
    return a, b

data_color = iio.imread('world_cup.jpg')
data = rgb2gray(data_color)

d1 = nd.geometric_transform(data, GeoFun)
plt.imshow(d1, cmap='gray')
plt.show()
```
